## Nodos, variables y supuestos específicos

### Mercado Total disponible
> <p style="text-align: justify"> El mercado total disponible es una constante obtenida a través de investigación. Se considera que el mercado objetivo son los estudiantes de estudios superiores. Dentro de los productos sustitutos se encuentran la gran variedad de chocolates, café y energizantes de forma general.</p>

### Mercado alcanzable por Canal
> <p style="text-align: justify"> Los canales por los cuales este producto llega a su consumidor son dos; Máquinas expededoras y tiendas de conveniencia. El desarrollo que siguen ambos canales es parecido en algunos nodos pero no igual. El mercado alcanzable, dado que no hay información y el estado temprano del negocio se propone modelar cuántos usuarios tienen acceso al producto con un distribución arbitraria.

> Como regla de operación, se espera que en un principio el mercado alcanzable sea bajo respecto al total potencial, dado a que es un producto de recién incorporación al mercado. Se considera que la distribución esté sesgada hacia la derecha ya que se cree más probable que el mercado tarde en incorporar el nuevo producto.

> Dentro de las opciones contempladas para la distribución del mercado alcanzable se encuentran aquellas que en su dominio se encuentran acotadas por un rango [a,b], ya que las maquinas de expendio tienen un límite de reabastecimiento; tambien se busca que puedan modelar un sesgo hacia la derecha.
 - Triangular
 - Beta 
 
> Se supone que las tiendas de conveniencia no tienen restricciones de abastecimiento se puede considerar una distribución en el rango $[a,\inf].$
- Erlang
  
> Para las máquinas expededoras se espera que el rango esté entre $[.01,.15]$.
 Para las tiendas se espera que con menos de .05 de probabilidad el mercado alcanzable se encuentre sobre .45 del total, con un mínimo de .01 de cobertura de mercado.

### Precio
> <p style="text-align: justify"> Al mercado que se busca entrar es aquel de golosinas y energizantes, mercado competido con gran variedad de productos sustitutos. Es por lo anterior que aun cuando se cuenta con un conponente de diferenciación sobre la competencia se prevee entrar a precios competitivos marcados por el mercado. El precio unitario se modelará constante.

### Ingresos
> <p style="text-align: justify">    
> Cabe resaltar antes de comenzar que las ventas es una variable discreta, de ser necesario se plantea discretizar distribuciones continuas para encontrar posibles soluciones.
   
> #### Máquinas    
> <p style="text-align: justify">    
> Las máquinas se encuentran en planteles educativos en lugares de tráfico peatonal. Hay un cierto número de unidades disponibles con oportunidad para crecer en el corto plazo; por el momento constantes. Dado el surtido semanal de las máquinas, hay un número limitado de unidades, por lo tanto el rango de ventas por máquina está limitado por el rango $[1,40]$ para periodos de un mes. De igual manera se propone una distribución beta, poisson, y  erlang bajo el supuesto que no hay límite práctico en la cantidad de producto en las máquinas[1].
    
>#### Tiendas de conveniencia
> <p style="text-align: justify">    
    > Los supuestos de la tienda de conveniencia son más relajdados, a falta de información se propondrá una distribución que parta de los parámteros de referencia que se puedan encontrar en información pública o reglas de negocio. Dentro de las posibles distribuciones se encuentran: poisson, Xi-cuadrada, beta, alguna otra gamma o gamma inversa. En el futuro, consolidándose el mercado cabe preguntarse si la distribución cambia a algo más 'normal'.      
    
> <small>[1] Este supuesto es factible dado que las maquinas tienen un límite de unidades por producto, pero se puede incrementar el número de filas de producto de la máquina si la demanda lo amerita; de igual manera se pueden incrementar las máquinas expendedoras en el corto plazo.</small>
