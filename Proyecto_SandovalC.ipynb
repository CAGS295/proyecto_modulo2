### Segundo Acercamiento: Estadístico $\chi$

Sea $f(x)$ la función de densidad empírica de $X$ y $\hat f(x)$ la densidad estimada consideremos el estadístico $\chi = [f(x) - \hat f(x)]^2, \forall x$

Si $\hat f(x) = f(x) \rightarrow \chi = 0,\chi \in [0,inf]$ 

<p style="text-align: justify">
Por lo tanto intentamos buscar la función de densidad que mejor se ajute a la distribución empírica minimizando $\chi$. Probando diferentes distribuciones continuas y utilizando el método de máxima verosimilitud ajustamos estimamos parámetros para funciones de densidad y calculamos el estadístico $\chi$ propuesto anteriormente para escoger la de mejor ajuste a la distribución empírica.

```python
def best_fit_distribution(data, bins=200):    
    """Model data by finding best fit distribution to data"""
    '''O. Jaramillo Zuloaga'''
    y, x = np.histogram(data, bins=bins, density=True)
    x = (x + np.roll(x, -1))[:-1] / 2.0
    
    from scipy import stats as st
    DISTRIBUTIONS = [        
        st.alpha,st.anglit,st.arcsine,st.beta,st.betaprime,st.bradford,st.burr,st.cauchy,st.chi,st.chi2,st.cosine,
        st.dgamma,st.dweibull,st.erlang,st.expon,st.exponnorm,st.exponweib,st.exponpow,st.f,st.fatiguelife,st.fisk,
        st.foldcauchy,st.foldnorm,st.frechet_r,st.frechet_l,st.genlogistic,st.genpareto,st.gennorm,st.genexpon,
        st.genextreme,st.gausshyper,st.gamma,st.gengamma,st.genhalflogistic,st.gilbrat,st.gompertz,st.gumbel_r,
        st.gumbel_l,st.halfcauchy,st.halflogistic,st.halfnorm,st.halfgennorm,st.hypsecant,st.invgamma,st.invgauss,
        st.invweibull,st.johnsonsb,st.johnsonsu,st.ksone,st.kstwobign,st.laplace,st.levy,st.levy_l,st.levy_stable,
        st.logistic,st.loggamma,st.loglaplace,st.lognorm,st.lomax,st.maxwell,st.mielke,st.nakagami,st.ncx2,st.ncf,
        st.nct,st.norm,st.pareto,st.pearson3,st.powerlaw,st.powerlognorm,st.powernorm,st.rdist,st.reciprocal,
        st.rayleigh,st.rice,st.recipinvgauss,st.semicircular,st.t,st.triang,st.truncexpon,st.truncnorm,st.tukeylambda,
        st.uniform,st.vonmises,st.vonmises_line,st.wald,st.weibull_min,st.weibull_max,st.wrapcauchy
    ]
    
    best_distribution = st.norm
    best_params = (0.0, 1.0)
    best_sse = np.inf    
    for distribution in DISTRIBUTIONS:
        import warnings        
        try:                
            warnings.filterwarnings('ignore')                
            params = distribution.fit(data)
            arg = params[:-2]
            loc = params[-2]
            scale = params[-1]
            pdf = distribution.pdf(x, loc=loc, scale=scale, *arg)
            sse = np.sum(np.power(y - pdf, 2.0))                                
            if best_sse > sse:
                best_distribution = distribution
                best_params = params
                best_sse = sse
        except Exception:
            pass 
    return (best_distribution.name, best_params)
```

## Proceso de difusión con salto

<p style="text-align:justify">
Los rendimientos del precio se parecen a una normal, excepto por las colas anchas. Se propone un proceso de difusión con salto para modelar los rendimientos del tipo de cambio libra peso con colas anchas. La justificación se debe a que dado el supuesto sobre la eficiencia del mercado, el mercado asimila la información nueva de manera muy rápida presentando los cambios abrúptos en el precio, esto es resultado de la asimetría de la información en los mercados.

Un proceso de poisson $N = N_t \in \mathbb Z_+,\forall t:[0,T]$ tiene intensidad \lambda y cumple con las siguientes propiedades:

- $N_0 = 0$
- $N_t - N_0 $ Sigue una distribución de poison: $P(N_t - N_0 = n) = e^{-\lambda t}\frac {\lambda t^n}{n!}$
- $N_t-N_s$ y $N_u-N_v$ son independientes si $(s,t) \cap (u,v) \ne 0$
- $N_{t+s}-N_s = N_t, \forall \, t,s > 0$ tiene incrementos estacionarios y en consecuencia $E[N_t] =Var(N_t)= \lambda t$

<p style="text-align:justify">
El proceso de poisson compensado $\hat N_t = N_t - \lambda t$ es una martingala y $E[\hat N_t] =Var(\hat N_t)= 0$ este hecho facilitará la estimación de los parámetros del proceso de difusión con salto ya que la media y la varianza de $\hat N_t$ no se mezclan con la tendencía ni la difusión del proceso completo.

Generador del Proceso $\mathcal{l}(y)$:

$\mathcal{L}_t^Yl(y) = \mathcal f_t \partial _y l(y) + \frac 12g_t^2 \partial _{yy}l(y) + \lambda([l(y +  h_t) - l(y)] - h_t\partial _y l(y))$

donde,
$Y_t = \int_0^t f_sds+\int_0^t g_sdW_s+\int_0^t h_{s^-} d\hat N_s$ es el proceso de difusión con salto

Tomando la solución general, $l(y)=Y_t$ simplica el generador a:

$\mathcal{L}_t^YY = \mathcal f_t \partial _y + \frac 12g_t^2 \partial _{yy} + \lambda([Y +  h_t - Y] - h_t\partial _y)$

El generador provee valiosa información del proceso estocástico. Indica el cambio esperado del proceso, formalmente como:


$$\mathcal L f(x) = \lim_{h \rightarrow 0} \frac {E[f(X_{t+h}\big | X_t=x)]-f(x)} {h}$$

```python
def poissonGen(mu):
    '''Generador de función de distribución de probabilidad'''
    k = 0
    prev = np.exp(-mu)   
    while True:        
        yield(prev)    
        k +=1
        prev = mu*prev/k
def CumPoisson(mu):
    '''Generador de función de distribución acumulada'''
    gen = poissonGen(mu)
    current = next(gen)
    while True:
        yield current
        current += next(gen)
```

<p style="text-align:justify">
La probabilidad de que no haya salto en el tiepo $t$ va decayendo exponencialmente a medida que se hace inminente la presencia de un salto, homólogamente la probabilidad de que se presente un salto va siendo cada vez más probable.
Utilizando la función de distribución de poisson con parámetro $\hat \lambda$ adaptado a la medida de probabilidad $\mathcal P$ de la filtración sobre la serie de tiempo del tipo cambiario GBPMXN, para el tiempo t generaremos un aleatorio $U \in [0,1]$ y compararemos con la distribución acumulada $F(x)$. Si $U < F(x | x=t-s)$ se presenta un salto y se empieza a tomar la probabilidad acumulada a partir de $s=t$.

Ejemplo: Con parámetro 5.4 para el proceso de poisson $N(t)$, salta con una esperanza de cada 5.4 intervalos de tiempo $t$.

```python
estimador = 5.4
generador = CumPoisson(estimador)
nojump = True
t =  0
s = t
r = np.random.rand()
while nojump:    
    p = next(generador)
    if r < p:
        nojump = False        
    print('t-s: {}\tU: {}\t P(s|s={})'.format(t-s,r,p))
    t +=1    
```

Corroboremos si realmente el tiempo medio de  espera entre saltos

```python
def instantes(est):    
    generador = CumPoisson(est)
    nojump = True
    t =  0
    r = np.random.rand()
    while nojump:                
        p = next(generador)
        if r < p:
            return t
        t +=1
```

<p style="text-align:justify">
Para encontrar los estimadores del proceso $Y_t = \int_0^t f_sds+\int_0^t g_sdW_s+\int_0^t h_{s^-} d\hat N_s$ se necesita aplicar la convolución de funciones de densidad para encontrar la densidad de $dY_t$ Dada las complicaciones de encontrar estimaciones para el proceso de salto se propone una aproximación al procesos de difusión con salto.

<p style='text-align: justify;'> Se puede mostrar que la distribución de $Y_t$ es una mezcla infinita de distribuciones gaussianas lo que hace que la estimación sea intratable; Una versión simplificada asumiendo que si la tasa de ocurrencia de saltos es pequeña, entonces en un período de tiempo suficientemente corto, solo puede ocurrir un salto. En este caso, $\Delta \hat N_t$ puede ser aproximado por una variable aleatoria de Bernoulli para $\lambda \Delta t$ pequeña, y la densidad de $ \Delta Y_t$ es

$$ f_{\Delta Y}(y) =(1 - \lambda \Delta t)\thinspace f_{\Delta D}(y) + \lambda \Delta t (f_{\Delta D} * f_{h_{s^-}})(y)$$

donde $f_{\Delta D}$ correponde a la densidad de $f_s\Delta s + g_s \Delta W_s$ que es el proceso de difusión con drift rate.

<p style='text-align: justify;'> Donde $f_{\Delta D}$ es la densidad de la parte de difusión y $f_{h_{s^-}}$ es la densidad asumida por los saltos. La convolución esta definida como $(f_{\Delta D} * f_{h_{s^-}})(y) = \int_\infty^\infty f_{\Delta D }(u)f_{h_{s^-}}(u-y)du$. La densidad de la difusión $f_{\Delta D}$ es una densidad normal con media $ (u- \frac{\sigma^2}{2})\Delta t + u_{h_{s^-}}$ y varianza $\sigma^2 \Delta t + \sigma^2_{h_{s^-}}$.

# Máxima verosimilitud logarítmica local

<p style="text-align:justify">
Se necesita encontrar los parámetros de la difusíon con salto ajustados a la filtración del tipo de cambio libra peso. Dado que encontrar la distribución de probabilidad del modelo implica convoluciones de un proceso de difusión con un proceso de poisson. Par encontrar los estimadores de máxima verosimilitud se necesita emplear convoluciones y ditribuciones producto.
Una vez obtenida la distribución de probabilidad de la convolución empleando el algoritmo de maximización de esperanza se busca maximizar iterativamente el logaritmo de la función de verosimilitud. El algoritmo de elección es un algoritmo iterativo heurístico, evolución diferencial.

```python
GP_r = GBPMXN.pct_change(1)[1:]
```

```python
from scipy.stats import norm
from scipy.optimize import differential_evolution as DE
import warnings
'lambda:0, mu:1, sigma:2, muq:3, sigmaq:4'
def density(theta,dy,dt):    
    mu1 = (theta[1] - theta[2]**2 / 2) * dt
    mu2 = mu1 + theta[3]
    sig1 = theta[2] * np.sqrt(dt)
    sig2 = np.sqrt(sig1**2 + theta[4]**2)
    pdf1 = norm.pdf(dy, mu1, sig1)
    pdf2 = norm.pdf(dy, mu2, sig2)    
    pdf = (1 - theta[0] * dt) * pdf1 + (theta[0] * dt) * pdf2
    return pdf 
```

```python
def negll(theta, dy, dt):    
    with warnings.catch_warnings():
        warnings.filterwarnings('ignore')
        L = density(theta, dy, dt)          
        nll = -sum(np.log(L))
        nll = 1e10 if np.isnan(nll) or np.isinf(nll) else nll
    return nll
```

```python
lims = [(0,1),(-1,1),(0,2),(-1,1),(0,2)]
```

> ### Estimación de parámetros

```python
res = DE(negll,bounds=lims,args=(GP_r,1),tol=.01,popsize=300)
```
